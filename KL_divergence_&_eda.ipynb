# Histogram Comparison 

```python
dfA =  ((df[df['Class_att'] == 'Abnormal']))
dfN =  ((df[df['Class_att'] == 'Normal']))
```

```python
dfA
```

```python
def uneven_kl_divergence(pk,qk):
    if len(pk)>len(qk):
        pk = np.random.choice(pk,len(qk))
    elif len(qk)>len(pk):
        qk = np.random.choice(qk,len(pk))
    return np.sum(pk * np.log(pk/qk))

#pelvic incidence
def hist(type_):
    
    kl_score = uneven_kl_divergence(dfA[type_],dfN[type_])
    plt.figure(figsize=(4,3))
    plot = plt.hist(dfA[type_], bins=100, alpha=0.5, label='Abnormal', density=True)
    plot = plt.hist(dfN[type_], bins=100, alpha=0.5, label='Normal', density=True)
    plt.xlabel('Data', size=9)
    plt.ylabel('Count', size=9)
    plt.title(f"Variable: {type_} \n KL: {kl_score}")
    plt.legend(loc='upper right')

```

```python
#https://datavizpyr.com/overlapping-histograms-with-matplotlib-in-python/
#add to end of chart segment (after plt.legent) this segment to save the image
    #plt.savefig("overlapping_histograms_with_matplotlib_Python_2.png")
hist('pelvic_incidence')
hist('pelvic_tilt')
hist('lumbar_lordosis_angle')
hist('sacral_slope')
hist('pelvic_radius')
hist('degree_spondylolisthesis')
hist('pelvic_slope')
hist('direct_tilt')
hist('thoracic_slope')
hist('cervical_tilt')
hist('sacrum_angle')
hist('scoliosis_slope')
```

```python
FEATURES = dfA.columns

kl_vals = [uneven_kl_divergence(dfA[f], dfN[f]) for f in FEATURES[:-2]]

```
